b). Do a scatter plot of weekofyear on x-axis and total_cases on y-axis and observe if there is a relationship.

```python
#Scatter plot with weekofyear in x-axis and total_cases on y-axis
merged.plot(kind="scatter",x="weekofyear",y="total_cases",color='red')
plt.title('Scatter plot of Number of cases each week')
plt.xlabel('WeekNumber')
plt.ylabel('NumberOfCases')
plt.show()
```

The number of cases seems to be less than hundred in most of the weeks throughout the year with an average around 25. The outliers seem to take place in the time period between 25 to 50 weeks suggesting the dengue spreads faster during the season between 25 to 50 weeks and so the number of cases seem to be high during that time period.

c). Do a box plot to observe the density of data and if it complies to Central Limit Theorem (for large enough data, mean of sample is mean of population, or data is tightly packed close to the mean)

```python
#Boxplot to observe the density of data
merged.boxplot(column='total_cases')
plt.title('Box plot of Number of cases')
plt.figure(figsize=(20,15))
plt.show()
```

The data is found to have a lots of outliers. The mean of the data and the mean of the population won't be close as the sample has many outliers in the week period between 25 to 50 weeks in terms of cases suggesting the data may not conform to a normal distribution and hence does not comply to Center Limit Theorem.

e). Read the relationship between mean and standard deviation from the link given here and describe the data of total_cases by city, if there is too much variation or too little variation https://www-users.york.ac.uk/~mb55/msc/applbio/week3/sd_text.pdf

```python
#Relation between mean and standard deviation for total_cases by both cities
total = merged.groupby(['city'])['total_cases']
total.hist(bins=50,color='orange')
plt.title('Histogram of total cases in both cities')
plt.xlabel('TotalCases')
plt.ylabel('Frequency')
plt.show()
```

```python
#Relation between mean and standard deviation for total_cases in city iq
total = merged[merged['city'] == 'iq'].groupby(['city'])['total_cases']
total.hist(bins=50,color='orange')
plt.title('Histogram of total cases in iq city')
plt.xlabel('TotalCases')
plt.ylabel('Frequency')
plt.show()
```

```python
#Relation between mean and standard deviation for total_cases in city sj
total = merged[merged['city'] == 'sj'].groupby(['city'])['total_cases']
total.hist(bins=50,color='orange')
plt.title('Histogram of total cases in sj city')
plt.xlabel('TotalCases')
plt.ylabel('Frequency')
plt.show()
```

The histogram of the total_cases by both cities shows us the data is right skewed and the city wise histogram also suggests the same. It means the data is not normally distributed and hence the relation between mean and standard deviation is too wide. The data suggests there is too much variance.

How many years in the city iq have greater than or equal to 50 total_cases of dengue and in which years?

```python
#How many years in city iq have equal to or more than 50 cases of dengue
df = merged[merged['city'] == 'iq'].groupby('year')['total_cases'].sum()
print(df[df>=50].count())
```

```python
#The years in which the number of cases were equal to or more than 50 in city iq
print(df[df>=50])
```
